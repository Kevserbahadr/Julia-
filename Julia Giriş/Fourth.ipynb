```julia
# 5) Değişkenlerinizin normal dağılıma uygunluğunu histogram ve t testi ile inceleyin.
```

```julia
using Pkg
Pkg.add("Plots")
Pkg.add("StatsBase")
```

```julia
Pkg.add("Distributions")
```

```julia
# Histogram ile normallik
```

```julia
using Distributions
using Plots

x = rand(1000)
d = fit(Normal, x)
myrandomvector = rand(d, 1000)

histogram(myrandomvector, nbins=20, fillalpha=0.3, label="fit")
histogram!(x, nbins=20, linecolor=:red, fillalpha=0.3, label="myvector")
```

```julia
# Histogram ile  küçük data normallik örneği 
```

```julia
using Distributions
using Plots

# Özel not değerleri
matematik_notlari = [75, 60, 80]

# Veriyi kullanarak özel bir dağılım oluştur
d = fit(DiscreteUniform, matematik_notlari)

# Dağılıma uygun rastgele notları oluştur
myrandomvector = rand(d, 1000)

# Histogramları çiz
histogram(myrandomvector, nbins=20, fillalpha=0.3, label="fit")
histogram!(matematik_notlari, nbins=20, linecolor=:red, fillalpha=0.3, label="myvector")
```

```julia
# Histogram ile normallik kendi datamız
```

```julia
x = st_age
d = fit(Normal, x)
myrandomvector = rand(d,1000)
histogram(myrandomvector,nbins=20,fillalpha=0.3)
histogram!(x,nbins=20,linecolor = :red,fillalpha=0.3)
```
